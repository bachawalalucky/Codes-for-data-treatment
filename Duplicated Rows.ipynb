# How do i find and remove duplicated rows in pandas?

```python
import pandas as pd
```

```python
df_cols=['age','gender','occupation','zip_code']
values=pd.read_table('http://bit.ly/movieusers',sep='|',header=None, names=df_cols)
values.head()
```

```python
values.head()
```

```python
#It shows that the total Number of duplicated rows i.e., entire repeated identical row present in the dataset 
values.duplicated().sum()
```

```python
#It shows that the total Number of duplicated rows i.e., repeated identical row present in the particular column 
values.zip_code.duplicated().sum()
```

```python
#this code is to know what are the identical rows present in a dataset
values.loc[values.duplicated(keep=False),:]
```

```python
#to drop duplicated rows
values.drop_duplicates()
```

```python
#to drop first duplicated rows
values.drop_duplicates(keep='first')
```

```python
 #to drop last duplicated rows
values.drop_duplicates(keep='last')
```

```python
#to drop all duplicated rows with simlar original rows
values.drop_duplicates(keep=False)
```

```python
#finally how to save data after removing duplicates
values=values.drop_duplicates(keep='last')
values
```
